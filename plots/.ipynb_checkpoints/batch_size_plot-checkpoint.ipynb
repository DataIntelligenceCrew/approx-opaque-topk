```python
import json
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
%config InlineBackend.figure_format = 'retina'
```

```python
# Raw Data
batch_sizes = [25, 50, 100, 200, 400, 600, 800]
scorer_times_secs = [0.03136857003785629, 0.02249977245467179, 0.01659403478590871, 0.013967893629168029, 0.0125180208041104, 0.012015344184915504, 0.0117932710461149]
memory_consumption_mb = [726.23, 1122.52, 1916.56, 3506.65, 6683.31, 9861.84, 13036.68]
max_gpu_memory = 20475.0
```

```python
# Calculations
scorer_times_millis = [x * 1000 for x in scorer_times_secs]
memory_percentage = [(mem / max_gpu_memory) * 100 for mem in memory_consumption_mb]

# Create the plot
fig, ax1 = plt.subplots()

# Plot scoring function latency
ax1.set_ylim([0.0, 34.0])
ax1.plot(batch_sizes, scorer_times_millis, color='black', label='Latency')
ax1.set_xlabel('Batch Size')
ax1.set_ylabel('Amortized Scoring Function Latency (ms)', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.grid(visible=True, color="#f0f0f0")

# Create a second y-axis for memory consumption
ax2 = ax1.twinx()
ax2.plot(batch_sizes, memory_percentage, color='black', linestyle='--', label='Memory Usage')
ax2.set_ylabel('GPU Memory Consumption (%)')
ax2.tick_params(axis='y', labelcolor='black')
ax2.set_ylim([0, max(memory_percentage) + 10])

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show plot
plt.title('Scoring Function Latency and Memory Usage')
plt.show()
```
